```python
# Proporción de compradores como 1 y no compradores como 0
hs["decision"].value_counts(normalize=True)
sns.countplot(x="decision", data=hs)
plt.title("Distribución de la variable objetivo (decision)")
plt.show()
```

Vemos un desbalance en 0 y 10 quiza mayor del 50%
- 0 No compra  y cuenta con aproximado de 150,000 registros
- 1 Si compra y cuenta con aproximado de 50,000 registros 


```python
# Lista de columnas que son texto
# No caracteristica predictiva
drop_cols = ["property_id"]
cat_cols = ["country", "city", "property_type", "furnishing_status"]
#El resto las trataremos como numéricas
num_cols = [col for col in hs.columns if col not in cat_cols + drop_cols + ["decision"]]
print(f"Columnas categóricas: {cat_cols}")
print(f"Columnas numéricas: {num_cols}")
```

```python
# Estadísticas descriptiva
hs.groupby("decision")[num_cols].agg(["mean", "median", "std"])    
```

Si las medias son muy diferentes las consideraremos para la predicción ( enfocandonos mas en satisfaction_score).
- Media en 0: 4.6
- Media en 1: 8.5
Observamos tambien que la desviación estandar  en compradores  es de 1.12 tal que el comprador queda satisfecho de forma consistente.
Variables price y property size (detalles esenciales para comprar una casa) pueden servir mas adelante, mientras la media es menor con los compradores.
emi to inconmecratio es el ratio de cuota mensual a ingresos, descrito como que porcentaje de tus ingresos mensuales brutos se destinan al pago de deudas
- D.E en 0: 0.242
- D.E en 1: 0.09

dando a entender que el comprador tiene un ratio de endeudamiento controlado, puede ser una condición para comprar


```python
cols_to_plot = ["price", "customer_salary", "monthly_expenses", "property_size_sqft", "loan_amount"]
fig, axes = plt.subplots(2,3, figsize=(15,10))
axes = axes.flatten()
for i, col in enumerate(cols_to_plot):
    for dec in [0,1]:
        axes[i].hist(hs[hs["decision"]==dec][col], bins=50, alpha=0.6, label=f"Decision={dec}")
    axes[i].set_title(col)
    axes[i].legend()
plt.tight_layout()
plt.show()
```

Distribuciones muy sesgadas, sebera trabajarse al entrenar la red.

```python
plt.figure(figsize=(15,10))
for i, col in enumerate(num_cols[:12]):
    plt.subplot(4,3,i+1)
    sns.boxplot(x="decision", y=col, data=hs)
    plt.title(f"Boxplot de {col} vs decisión")
plt.tight_layout()
plt.show()
```

Podemos ver que hay outliers y diferencias de la mediana.

```python
plt.figure(figsize=(14,10))
corr = hs[num_cols + ["decision"]].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Matriz de correlación")
plt.show()
```

```python
fig, axes = plt.subplots(2,2, figsize=(15,13))
for i, col in enumerate(cat_cols):
    # Porcentaje de compra por categoria
    prof_hs = hs.groupby(col)["decision"].mean().sort_values(ascending=False)
    sns.barplot(x=prof_hs.values, y=prof_hs.index, ax=axes[i//2, i%2])
plt.tight_layout()
plt.show()
```
